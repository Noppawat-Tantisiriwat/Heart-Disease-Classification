# EDA : Explordatory Data Analysis

```python
# load data
data = pd.read_csv("heart.csv")
```

```python
data
```

```python
# data discription
d_description = data.describe()
```

```python
# features and labels
X, y = data.drop(["target"], axis=1), data[["target"]]
```

```python
# print features
X
```

```python
# print labels
y
```

```python
# split datasets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
categorical_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
```

```python
d_num = d_description.drop(categorical_cols, axis=1)
```

```python
X_train_num, X_test_num = X_train.drop(categorical_cols, axis=1), X_test.drop(categorical_cols, axis=1)
```

```python
X_train_num
```

```python
for col in X_train_num.columns:
    mu = d_num[col]["mean"]
    sigma = d_num[col]["std"]
    X_train_num[col] = X_train_num[col].apply(lambda x: (x - mu) / sigma)


for col in X_test_num.columns:
    mu = d_num[col]["mean"]
    sigma = d_num[col]["std"]
    X_test_num[col] = X_test_num[col].apply(lambda x: (x - mu) / sigma)
```

```python
X_train_num_normalized = X_train_num
```

```python
X_test_num_normalized = X_test_num
```

```python
X_train_cat, X_test_cat = X_train[categorical_cols], X_test[categorical_cols]
```

```python
X_train_final = pd.merge(X_train_cat, X_train_num, left_index=True, right_index=True)
```

```python
X_test_final = pd.merge(X_test_cat, X_test_num, left_index=True, right_index=True)
```

```python
heart_train = pd.merge(X_train_final, y_train, left_index=True, right_index=True)
```

```python
heart_test = pd.merge(X_test_final, y_test, left_index=True, right_index=True)
```

```python
heart_train.to_csv("heart_train.csv")
```

```python
heart_test.to_csv("heart_test.csv")
```
